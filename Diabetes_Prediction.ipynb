Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and analysis

```python
# loading the dataset "diabetes" into a pandas dataframe
diabetes_dataset = pd.read_csv("/content/Diabetes.csv", header=None)
```

Printing the first 5 rows of the dataset

```python
diabetes_dataset.head()
```

```python
# number of rows and columns in this dataset
diabetes_dataset.shape
```

```python
# getting the statistical measures of the data
diabetes_dataset.describe()
```

Finding the number of diabetic and non-diabetic cases

```python
diabetes_dataset[8].value_counts()
```

Getting mean of diabetic and non-diabetic cases

```python
diabetes_dataset.groupby(8).mean()
```

Separating the data and the labels


```python
X = diabetes_dataset.drop(columns=8, axis=1)
Y = diabetes_dataset[8]
```

```python
X.head()
```

```python
Y.head()
```

Data Standardization (because values of 'glucose', 'age', etc. are not in the same range)

```python
# Making an object of the StandardScaler that we just imported
scaler = StandardScaler()
```

```python
# Loading the scaler object with our data.. it analyzes the data and calculates a "Transforming value" to be used later
scaler.fit(X)
```

```python
# actual transformation of the data
std_data = scaler.transform(X)

pd.DataFrame(std_data).head()
```

Train-Test data split

```python
X_train, X_test, Y_train, Y_test = train_test_split(std_data, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

Creating the model and training it

```python
# creating the model object
classifier = svm.SVC(kernel='linear')
```

```python
# training the svm classifier
classifier.fit(X_train, Y_train)
```

Evaluating the model (on training data)


```python
# accuracy score of the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
# printing the accuracy score
print("Accuracy score of the model: ", training_data_accuracy)
```

Testing the model (on testing data)

```python
# testing the model
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# printing the result
print("Accuracy on test data: ", test_data_accuracy)
```

Making a predictive system

```python
input_data = (8,133,72,0,0,32.9,0.27,39) # 1: diabetic

# converting input_data to a numpy array
input_data_as_array = np.asarray(input_data)

# reshape the above array: model was trained on a 2D array (on 768 samples), but now we need to test on only 1 sample, i.e. (we need to test it on 1D array)
input_data_reshaped = input_data_as_array.reshape(1, -1)

# now standardize this also... just like you did during the training
input_data_standardized = scaler.transform(input_data_reshaped)  # now scaler.fit() not needed again; it's needed only at the start

# now predict
prediction = classifier.predict(input_data_standardized)

# print the prediction
print("Prediction based on the given data is:", "Diabetic" if prediction[0]==1 else "Healthy")
```
